```python
def sigmoid(z):
    return (1 / (1 + np.exp(-z)))

print(sigmoid(0.0001))
print(sigmoid(3))
print(sigmoid(-100.002))
```

```python
def predict(X, theta):
      return(sigmoid((X * theta)))
```

```python
def cost(X, y, theta):
    return((-1 / X.shape[0]) * np.sum(y * np.log(predict(X, theta)) + (1 - y) * np.log(1 - predict(X, theta))))
```

```python
X = np.c_[np.ones(X.shape[0]), X]
print (X)
```

```python
theta = np.zeros(3, dtype=np.float64)
cost(X, y, theta)
```

```python
from tqdm import tqdm
def fit(X, y, theta, alpha, num_iters):
    # Initialiser certaines variables utiles
    m = X.shape[0]
    J_history = []
    for _ in tqdm(range(num_iters)):
        theta = theta - (alpha / m) * (np.dot(predict(X, theta) - y, X))
        J_history.append(cost(X, y, theta))
    return theta, J_history
```

```python
theta = np.zeros(3, dtype=float)
theta, J_history = fit(X, y, theta, 0.001547, 6000000)
```
